```python
areas = []
for i in range(len(df)):
    if (df.iloc[i]['total_area_sq_km'] is None) or (df.iloc[i]['total_area_sq_km']== 0):
        areas.append(i)
    if (df.iloc[i]['forest_area_sqkm'] is None) or (df.iloc[i]['forest_area_sqkm']== 0):
        areas.append(i)
df.drop(areas)
```
